# Import the Librarys

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Read Data

```python
df = pd.read_csv('dirty_supermarket.csv')
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df.sort_values('Branch', ascending=True, inplace = True)
df.head()
```

```python
df.tail()
```

```python
df.to_csv('clean_supermarket.csv')
```

# Visualization

# Best selling product lines

```python
plt.figure(figsize=(8,6))
sns.barplot(x=df.index, y=df['Product line'])
```

# Total outlier

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Total'])
plt.show()
```

# Which payment method more used

```python
plt.figure(figsize=(6,4))
plt.hist(df['Payment'],bins=10,color='red')
plt.show()
```

# Heatmap correlations

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(),annot=True)
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Product line',data=df)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(df['Quantity'],df['Rating'],df['Total'])
ax.set_xlabel('Quantity')
ax.set_ylabel('Rating')
ax.set_zlabel('Total')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x='Quantity',data=df)
```
